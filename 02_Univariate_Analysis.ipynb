### <a id="c1_ID"> 3.1 Box Plots</a>

Box plots displays the distribution of numerical data based on the five number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The ends of the box are the first and third quartiles. The horizontal line in the middle of the box represents the median. The whiskers of the boxplot are the two horizontal lines outside of the box and represent the minimum and maximum observations. Boxplots can also be used to detect outliers. Outliers are the observations found outside the whiskers of the boxplot, outside 1.5IQR (IQR = Q3 - Q1) above the third quartile or below the first quartile.

```python
# Box plots for all numeric features
fig = plt.figure(figsize=(6,21))

ax1 = plt.subplot(5, 1, 1)
sns.boxplot(data = vgsales, y = "NA_Sales", ax=ax1)

ax2 = plt.subplot(5, 1, 2)
sns.boxplot(data = vgsales, y = "EU_Sales", ax=ax2)

ax3 = plt.subplot(5, 1, 3)
sns.boxplot(data = vgsales, y = "JP_Sales", ax=ax3)

ax4 = plt.subplot(5, 1, 4)
sns.boxplot(data = vgsales, y = "Other_Sales", ax=ax4)

ax5 = plt.subplot(5, 1, 5)
sns.boxplot(data = vgsales, y = "Global_Sales", ax=ax5)

plt.show()
```

From the boxplots, we can see the presence of outliers. Outliers tend to have more significant weight and have a negative impact on data analysis if they are not treated properly for small datasets. Since our dataset is large and outliers are expected in our sales data as these indicate video games with very good sales, these are unlikely to be errors in the data, so we will keep these outliers for our analysis.

### <a id="c2_ID"> 3.2 Histograms</a>

Histograms display the distribution of numerical data. It is an estimate of the probability distribution of continuous variables. They can have normal, right-skewed (positively skewed), or left-skewed (negatively skewed) distribution.

```python
# Histograms for all numeric features
fig = plt.figure(figsize=(6,21))

ax1 = plt.subplot(5, 1, 1)
sns.distplot(vgsales["NA_Sales"], kde = False, rug = True, ax=ax1)

ax2 = plt.subplot(5, 1, 2)
sns.distplot(vgsales["EU_Sales"], kde = False, rug = True, ax=ax2)

ax3 = plt.subplot(5, 1, 3)
sns.distplot(vgsales["JP_Sales"], kde = False, rug = True, ax=ax3)

ax4 = plt.subplot(5, 1, 4)
sns.distplot(vgsales["Other_Sales"], kde = False, rug = True, ax=ax4)

ax5 = plt.subplot(5, 1, 5)
sns.distplot(vgsales["Global_Sales"], kde = False, rug = True, ax=ax5)

plt.show()
```

Because this is a large dataset, we will not be using Shapiro test to test for normality. The histograms for all the numeric features are right-skewed. The distributions for NA_Sales and Other_Sales look similar. JP_Sales has a less narrow distribution. Since all of the numeric features are not normally distributed, we will need to perform non-paramteric tests for our analysis.

### <a id="c3_ID">3.3 Countplots</a>

Countplots display the distribution of categorical data.

```python
# Determine Value Counts for "Platform" column
vgsales.Platform.value_counts()
```

```python
# Countplot for Platform
plt.figure(figsize=(15,10))
sns.countplot(x = "Platform", data = vgsales)
plt.xticks(rotation=45)
plt.show()
```

From the Platform countplot, we can see a higher number of video game sales from DS and PS2 and lower number of video game sales from SCD, WS, and 3D0.

```python
# Determine Value Counts for "Year" column
vgsales.Year.value_counts()
```

```python
# Countplot for Year
plt.figure(figsize=(15,10))
sns.countplot(x = "Year", data = vgsales)
plt.xticks(rotation=45)
plt.show()
```

From the Year countplot, we can see a higher number of video game sales from 2008 and 2009 and lower number of video game sales from 1980 to 1990. It is also interesting to note the normal distribution about the years 2008 and 2009. 0 is indictive of missing values.

```python
# Determine Value Counts for "Genre" column
vgsales.Genre.value_counts()
```

```python
# Countplot for Genre
plt.figure(figsize=(15,10))
sns.countplot(x = "Genre", data = vgsales)
plt.xticks(rotation=45)
plt.show()
```

From the Genre countplot, we can see a higher number of video game sales from Action and Sports and lower number of video game sales from Puzzle, Strategy, Simulation, and Fighting categories.
