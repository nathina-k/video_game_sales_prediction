```python
# Dataframe showing evaluation metric values, evaluation metrics used, and machine learning models trained and tested
metrics_df = pd.DataFrame(metric_values, columns = columns, index = index)
metrics_df
```

```python
# Heatmap showing evaluation metric values, evaluation metrics used, and machine learning models trained and tested
plt.figure(figsize=(20,12))
sns.heatmap(metrics_df, annot=True, fmt='.5f', cbar=True,
            xticklabels = columns, cmap="coolwarm", linecolor='black', linewidths=1,
            yticklabels = index)
plt.xlabel('Evaluation Metrics')
plt.ylabel('Regression Models')
plt.show()
```

To select the best model for predicting North American video games sales, mean absolute error, mean squared error, and root mean squared error have to be very low and R-squared which captures the variance in the data has to be very high. Therefore, the best performing model for our dataset is Multilinear Regression, followed by Multilinear Regression + RFE, Multilinear Regression + PCA, KNN Regression, KNN Regression + RFE, and KNN Regression + PCA. The SVM regression models did not perform well on our dataset.

In conclusion, after features were scaled, regression models were built without RFE or PCA, with RFE, and with PCA. Regression models included Multilinear regression, KNN regression, and SVM regression. Of these nine models, Multilinear regression was the best  performeing model for predicting North American video game sales with a Mean Absolute Error of 0.18, Mean Squared Error of 0.22, Root Mean Squared Error of 0.46, and R-squared of 0.76. 
For future work, for RFE, linear rgression was used in our analysis as it was more time efficient. Looking into SVM regression could be helpful because in our analysis, RFE did not give better results. Also, other feature selection methods can be explored like filter-based or hybrid-based methods. We can conclude that PCA did not perform as well as RFE.
